```python
import pandas as pd
import sqlite3
```

```python
#read the CSV file into a Pandas DataFrame
csv_file_path = './Music_metrix.csv'
country_music = pd.read_csv(csv_file_path)
country_music
```

```python
#create a sqlite database connection
country_sqlite = 'music_in_countries.db'
sqlite_conn = sqlite3.connect(country_sqlite)
 
```

```python
created_table = 'countries_music'
country_music.to_sql(created_table, sqlite_conn, if_exists='replace', index=False)
```

```python
#connect to the sqlite database
cursor = sqlite_conn.cursor()
```

```python
#Query the data
cursor.execute("SELECT * FROM countries_music")
rows = cursor.fetchall()
```

```python
#Get Column names
column_names = [description[0] for description in cursor.description]
```

```python
#Convert the data to a list of dictionaries
data = []
for row in rows:
    data.append(dict(zip(column_names,row)))
```

```python
#Write the data to a JSON file
import json

with open('output.json', 'w') as json_file:
    json.dump(data, json_file, indent=4)
```
